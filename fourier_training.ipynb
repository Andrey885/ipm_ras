```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
```

```python
brightness = np.load('arr.npy')
fps = 25
smooth = 5
time = np.linspace(0,len(brightness), num = len(brightness))/fps
yhat = savgol_filter(brightness, smooth, 3)#3 for polynom order
plt.plot(time, yhat)
plt.grid()
plt.xlabel('Time, seconds')
plt.ylabel('Mean pixel')
plt.show()
### fourier
#time = np.asarray(time[1:len(time)])
#time = time[1:len(time)]
time = np.linspace(fps/len(brightness), fps,len(brightness))
#brightness = brightness[1:len(brightness)]
#yhat = yhat[1:len(yhat)]
yf = scipy.fftpack.fft(yhat)

plt.plot(time, np.abs(yf))
plt.grid()
plt.xlabel('Frequency')
plt.ylabel('Mean pixel fourier')
plt.savefig('.\plots\S39_sSi+QDs_growth_plot'+ 'jupyter'+ '.jpg')
plt.show()
```
